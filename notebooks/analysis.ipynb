Does Toss Matter More in Certain Venues?

```python
df['toss_win_match_win'] = df['toss_winner'] == df['winner']
venue_toss = df.groupby('venue')['toss_win_match_win'].mean() * 100
top_venues = venue_toss.sort_values(ascending=False).head(10)
top_venues.sort_values().plot(kind='barh', figsize=(12,6))

plt.title("Toss Impact by Venue (%)")
plt.xlabel("Win % when Toss is Won")
plt.ylabel("Venue")

plt.show()
```

🔍 Insight:

- Certain venues show a higher percentage of matches where the toss winner also wins the match.
- This suggests that toss advantage is more impactful in specific stadiums.
- Pitch conditions and ground size may influence match outcomes after winning the toss.

At each stadium, which team wins the most?

```python
venue_team = df.groupby(['venue','winner']).size().reset_index(name='wins')
top_teams = venue_team.sort_values(['venue','wins'], ascending=[True, False])

top_teams = top_teams.groupby('venue').head(1)
top_teams.head(10)
```

```python
top_teams_sorted = top_teams.sort_values(by='wins', ascending=False).head(10)
top_teams_sorted['label'] = top_teams_sorted['winner'] + " @ " + top_teams_sorted['venue']
top_teams_sorted.plot(
    x='label',
    y='wins',
    kind='barh',
    figsize=(12,6)
)

plt.title("Top Teams Dominating Specific Venues")
plt.xlabel("Number of Wins")
plt.ylabel("Team @ Venue")

plt.tight_layout()
plt.show()
```

🔍 Insight:

- Certain teams dominate specific venues consistently.
- This highlights strong home-ground advantage.
- For example, teams like CSK at Chepauk and MI at Wankhede perform significantly better.
