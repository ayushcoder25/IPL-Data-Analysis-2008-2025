```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Importing two Datasets and Merging together

```python
df = pd.read_csv("IPLData.csv")
df_2025 = pd.read_csv("IPL25.csv")
df_2025 = df_2025.rename(columns={
    'match_id': 'id',
    'match_winner': 'winner',
    'player_of_the_match': 'player_of_match'
})
df_2025 = df_2025[[
    'id', 'date', 'venue', 'team1', 'team2',
    'toss_winner', 'toss_decision', 'winner',
    'player_of_match'
]]
missing_cols = set(df.columns) - set(df_2025.columns)

for col in missing_cols:
    df_2025[col] = None
    
df_2025 = df_2025[df.columns]

df_final = pd.concat([df, df_2025], ignore_index=True)
df_final
```

Fixing team names by Mapping 


```python
print(df_2025['team1'].unique())
```

```python
team_map = {
    "MI": "Mumbai Indians",
    "CSK": "Chennai Super Kings",
    "RCB": "Royal Challengers Bangalore",
    "KKR": "Kolkata Knight Riders",
    "SRH": "Sunrisers Hyderabad",
    "DC": "Delhi Capitals",
    "RR": "Rajasthan Royals",
    "PBKS": "Kings XI Punjab",   
    "GT": "Gujarat Titans",
    "LSG": "Lucknow Super Giants"
}

df_2025['team1'] = df_2025['team1'].map(team_map)
df_2025['team2'] = df_2025['team2'].map(team_map)
df_2025['winner'] = df_2025['winner'].map(team_map)
df_2025['toss_winner'] = df_2025['toss_winner'].map(team_map)
print(df_2025[['team1','team2','winner']].head())
```

Fixing names

```python
player_map = {
    "V Kohli": "Virat Kohli",
    "RG Sharma": "Rohit Sharma",
    "DA Warner": "David Warner",
    "AD Russell": "Andre Russell",
    "CH Gayle": "Chris Gayle"
}
top_old = df['player_of_match'].value_counts().head(20)
df['player_of_match'] = df['player_of_match'].replace(player_map)
df['date'] = pd.to_datetime(df['date'])
df_2025['date'] = pd.to_datetime(df_2025['date'], format='%B %d,%Y')
df_final = pd.concat([df, df_2025], ignore_index=True)
df_final
```

```python
df_final.to_csv("ipl_cleaned.csv", index=False)
```
